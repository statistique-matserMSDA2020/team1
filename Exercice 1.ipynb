```python
import numpy as np
import matplotlib.pyplot as plt
import math
import csv
import seaborn as sns
```

# 1. Simulation d'un échantillon de taille 1000 suivant la loi binômiale B(30, 0.2)

```python
B = np.random.binomial(n=10, p=0.5, size=10000)
print(B)
plt.figure(figsize=(10,8))
plt.hist(B)
plt.show()
```

# 2. Simulation d'un échantillon de taille 1000 suivant la loi normale N(3, 4)

```python
N = np.random.normal(3, 4, size=10000)
print(N)
plt.figure(figsize=(10,8))
sns.distplot(N, hist=False)
plt.show()
```

```python
G = np.random.gamma(10,5, size=10000)
print(G)
plt.figure(figsize=(10,8))
sns.distplot(G, hist=False)
plt.show()
```
